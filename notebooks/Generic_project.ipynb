#### Correlations between features



```python
# Correlation between features

correlation_matrix = data.corr()

# Converting the correlation matrix into a DataFrame
corr_df = correlation_matrix.stack().reset_index()

# Renaming the columns
corr_df.columns = ['feature_1','feature_2', 'correlation']

# Remove "self correlations"
no_self_correlation = (corr_df['feature_1'] != corr_df['feature_2'])
corr_df = corr_df[no_self_correlation]

corr_df['absolute_correlation'] = np.abs(corr_df['correlation'])

corr_df.sort_values(by="absolute_correlation", ascending=False, inplace = True)
corr_df
```

## Missing values


```python
# Counting the percentage of NaN for each column
#NaN percentage for each column
missing_data = data.isnull().sum().sort_values(ascending=False)/len(data)*100 

# missing_data
# print("Missing data in percentage : ");

missing_data[missing_data > 0];
```

```python
#Create a df showing the percentage of missing data
missing_values_df = pd.DataFrame(missing_data, columns = ['Percent_missing'])
missing_values_df = missing_values_df[missing_values_df['Percent_missing'] > 0]

missing_values_df = missing_values_df.reset_index()\
                    .rename(columns = {'index' : 'features'})

# Missing values columns (with percentage of missing values > 0)
missing_values_columns = missing_values_df['features'].values.tolist()

#Retrieve types of columns with missing data as a Dataframe
type_missing_values_df = pd.DataFrame(data[missing_values_columns].dtypes, columns=['Type'])
type_missing_values_df = type_missing_values_df.reset_index()\
                        .rename(columns = {'index' : 'features'})

#Merge features, feature type and percentage of missing values
missing_values_df = pd.merge(missing_values_df, type_missing_values_df)
missing_values_df
```

```python
#Missing value columns
print(f"Length of features with missing values : {len(missing_values_columns)}\n")
print(f"columns with values missing : {missing_values_columns}")
```

### Missing values : Categorical columns


```python
#Categorical features with missing values
categorical_feat_missing = missing_values_df[missing_values_df.Type == 'object']['features'].values.tolist()

print(f"Length of Categorical features with missing values : {len(categorical_feat_missing)}\n")
print(f"Remaining categorical columns with values missing : {categorical_feat_missing}")
```
